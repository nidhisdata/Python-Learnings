```python
import pandas as pd
```

```python
df = pd.read_csv('/content/rows.csv')
df.head()
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df['Date Recorded'] = pd.to_datetime(df['Date Recorded'],format = '%m/%d/%Y')
```

```python
df['day'] = df['Date Recorded'].dt.day
df['month'] = df['Date Recorded'].dt.month
df['year'] = df['Date Recorded'].dt.year
display(df)
```

```python
print(df.describe())
```

```python
df['Sale Amount'] = pd.to_numeric(df['Sale Amount'], errors='coerce')
```

```python
aggregated_data = df.groupby(['year']).agg(
    sum_sales=('Sale Amount', 'sum'),
    mean_sales=('Sale Amount', 'mean'),
    median_sales=('Sale Amount', 'median'),
    max_sales=('Sale Amount', 'max'),
    min_sales=('Sale Amount', 'min')
)

display(aggregated_data)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Corrected: Providing 'height' argument and removing unnecessary 'bins'
plt.bar(x=aggregated_data.index.get_level_values('year'), height=aggregated_data['sum_sales'], edgecolor='black')

# Adding title and labels
plt.title('Distribution of Sale Amount')
plt.xlabel('Month')  # Changed x-axis label to 'Month'
plt.ylabel('Average Sales')  # Changed y-axis label to 'Average Sales'

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Group data by year and residential type, calculate mean sales
grouped_data = df.groupby(['year', 'Residential Type'])['Sale Amount'].mean().reset_index()

# Get unique years and residential types
years = grouped_data['year'].unique()
residential_types = grouped_data['Residential Type'].unique()

# Set up the bar positions
bar_width = 0.35
index = np.arange(len(years))

# Create the clustered bar chart
fig, ax = plt.subplots(figsize=(30, 6))

for i, res_type in enumerate(residential_types):
    # Filter data for the current residential type
    data = grouped_data[grouped_data['Residential Type'] == res_type]

    # Ensure that 'data' contains values for all years in 'index'
    # If a year is missing for the current residential type, it will be filled with 0
    data = data.set_index('year').reindex(years, fill_value=0)

    ax.bar(index + i * bar_width, data['Sale Amount'], bar_width, label=res_type)

# Set labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Mean Sale Amount')
ax.set_title('Mean Sale Amount by Year and Residential Type')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(years)
ax.legend()

plt.show()
```

```python
df['difference'] = df['Sale Amount'] - df['Assessed Value']
```

```python
import matplotlib.pyplot as plt

difference_by_year = df.groupby('year')['difference'].mean()

# Get the x-values (years)
x = difference_by_year.index

# Get the y-values (mean difference for each year)
y = difference_by_year.values

# Creating the line plot
plt.plot(x, y, marker='o')

# Adding title and labels
plt.title('Line Plot of Mean Difference between Sale Amount and Assessed Value by Year')
plt.xlabel('Year')
plt.ylabel('Mean Difference')

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


difference_by_year = df.groupby('year')['difference'].mean()
standard_deviation = df.groupby('year')['difference'].mean().std()

# Get the x-values (years)
x = difference_by_year.index

# Get the y-values (mean difference for each year)
y = difference_by_year.values

# Creating the line plot
plt.plot(x, y, marker='o')

x_values = np.array(x)
y_values = np.array(y)

n = len(x_values)
x_sum = np.sum(x_values)
y_sum = np.sum(y_values)
xy_sum = np.sum(x_values * y_values)
x_squared_sum = np.sum(x_values ** 2)

m = (n * xy_sum - x_sum * y_sum) / (n * x_squared_sum - x_sum ** 2)
b = (y_sum - m * x_sum) / n

y_line = m * x_values + b

#plt.plot(x, y + standard_deviation, color='red', linestyle='--')
plt.plot(x, y - standard_deviation, color='red', linestyle='--')
plt.plot(x, y_line, color='blue', linestyle='--')

# Adding title and labels
plt.title('Line Plot of Mean Difference between Sale Amount and Assessed Value by Year')
plt.xlabel('Year')
plt.ylabel('Mean Difference')

# Displaying the plot
plt.show()
```
